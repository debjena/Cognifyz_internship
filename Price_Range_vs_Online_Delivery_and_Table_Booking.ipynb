```python
# üì¶ Import and load dataset
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset .csv")

# ‚úÖ Step 1: Clean necessary columns
df['Price range'] = pd.to_numeric(df['Price range'], errors='coerce')
df['Has Online delivery'] = df['Has Online delivery'].fillna('No')
df['Has Table booking'] = df['Has Table booking'].fillna('No')

# ‚úÖ Step 2: Analyze Online Delivery by Price Range
online_delivery = df.groupby(['Price range', 'Has Online delivery']).size().unstack(fill_value=0)
online_delivery['Total'] = online_delivery.sum(axis=1)
online_delivery['% with Online Delivery'] = round((online_delivery['Yes'] / online_delivery['Total']) * 100, 2)

# ‚úÖ Step 3: Analyze Table Booking by Price Range
table_booking = df.groupby(['Price range', 'Has Table booking']).size().unstack(fill_value=0)
table_booking['Total'] = table_booking.sum(axis=1)
table_booking['% with Table Booking'] = round((table_booking['Yes'] / table_booking['Total']) * 100, 2)

# ‚úÖ Step 4: Print results
print("üì¶ Online Delivery by Price Range:")
display(online_delivery[['Total', '% with Online Delivery']])

print("\nüçΩÔ∏è Table Booking by Price Range:")
display(table_booking[['Total', '% with Table Booking']])

# ‚úÖ Step 5: Visualization

# Bar Plot - Online Delivery
plt.figure(figsize=(8, 4))
plt.bar(online_delivery.index, online_delivery['% with Online Delivery'], color='skyblue')
plt.title('% Restaurants Offering Online Delivery by Price Range')
plt.xlabel('Price Range (1=Low, 4=High)')
plt.ylabel('Percentage (%)')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.show()

# Bar Plot - Table Booking
plt.figure(figsize=(8, 4))
plt.bar(table_booking.index, table_booking['% with Table Booking'], color='salmon')
plt.title('% Restaurants Offering Table Booking by Price Range')
plt.xlabel('Price Range (1=Low, 4=High)')
plt.ylabel('Percentage (%)')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.show()
```
