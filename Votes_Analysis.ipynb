```python
# üì¶ Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# ‚úÖ Load dataset (Upload "Dataset .csv" manually if not already loaded)
df = pd.read_csv("Dataset .csv")

# ‚úÖ Clean and prepare votes and rating columns
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')

# Filter out rows with missing or 0 votes
df_votes = df.dropna(subset=['Votes', 'Aggregate rating'])
df_votes = df_votes[df_votes['Votes'] > 0]

# ‚úÖ Step 1: Highest and Lowest Voted Restaurants

max_votes = df_votes['Votes'].max()
min_votes = df_votes['Votes'].min()

highest_voted = df_votes[df_votes['Votes'] == max_votes][['Restaurant Name', 'Votes', 'Aggregate rating']]
lowest_voted = df_votes[df_votes['Votes'] == min_votes][['Restaurant Name', 'Votes', 'Aggregate rating']]

print("üèÜ Highest Voted Restaurant(s):")
display(highest_voted)

print("\nüßæ Lowest Voted Restaurant(s):")
display(lowest_voted)

# ‚úÖ Step 2: Correlation between Votes and Rating

correlation = df_votes[['Votes', 'Aggregate rating']].corr().iloc[0, 1]
print(f"\nüìà Correlation between votes and rating: {round(correlation, 3)}")

# ‚úÖ Step 3: Visualization - Scatter Plot

plt.figure(figsize=(8, 5))
plt.scatter(df_votes['Votes'], df_votes['Aggregate rating'], alpha=0.5, color='teal')
plt.title('Votes vs Aggregate Rating')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.grid(True)
plt.show()
```
