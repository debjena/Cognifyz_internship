```python
# Task 1: Top Cuisines

import pandas as pd
import matplotlib.pyplot as plt

# Upload your file manually in Colab (from left panel)
df = pd.read_csv('Dataset .csv')

# Drop missing cuisine values
cuisine_series = df['Cuisines'].dropna()

# Split all cuisines into a flat list
all_cuisines = [c.strip() for sublist in cuisine_series.str.split(',') for c in sublist]

# Count cuisine frequency
from collections import Counter
cuisine_counts = Counter(all_cuisines)

# Top 3 cuisines
top_3 = cuisine_counts.most_common(3)
print("Top 3 Cuisines:", top_3)

# Calculate percentage
total_restaurants = df.shape[0]
percentages = {}
for cuisine, _ in top_3:
    count = df['Cuisines'].dropna().apply(lambda x: cuisine in x).sum()
    percentages[cuisine] = round((count / total_restaurants) * 100, 2)

# Display results
for cuisine, percent in percentages.items():
    print(f"{cuisine}: {percent}% of restaurants")

# Bar chart
plt.figure(figsize=(8, 5))
plt.bar(percentages.keys(), percentages.values(), color='orange')
plt.title("Top 3 Most Common Cuisines (Percentage)")
plt.ylabel("Percentage of Restaurants")
plt.xlabel("Cuisine")
plt.grid(axis='y')
plt.show()
```
